```python
salary_by_currency = df_selected.dropna(subset=['Currency', 'ConvertedCompYearly', 'CompTotal'])
salary_in_usa = salary_by_currency[(salary_by_currency['Currency'] == 'USD\tUnited States dollar') & (salary_by_currency['ConvertedCompYearly'] > 100) & (salary_by_currency['ConvertedCompYearly'] < 2500000)]
salary_in_usa.sort_values(by='ConvertedCompYearly', ascending=False, inplace=True)
salary_in_usa.loc[salary_in_usa['ConvertedCompYearly'] == salary_in_usa['ConvertedCompYearly'].max()]
```

```python
import matplotlib.ticker as ticker

def custom_format(x, pos):
    return f'{int(x):,}'

```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(data=salary_in_usa, y='ConvertedCompYearly', x='Age', hue='EdLevel', )
plt.xticks(rotation=45)
plt.xlabel('Age')
plt.ylabel('Salary in $')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(custom_format))
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
salary_usa_pro_dev = salary_in_usa.dropna(subset=['YearsCode', 'YearsCodePro',])
salary_usa_pro_dev.info()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.set_theme(style='white', context='talk')
sns.barplot(data=salary_usa_pro_dev[:20], y='ConvertedCompYearly', x='LanguageHaveWorkedWith', hue='YearsCodePro', palette='pastel', orient='v')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(custom_format))
plt.xticks(rotation=90, size=8)
ax.legend(title='YearsCodePro', loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('LanguageHaveWorkedWith')
plt.ylabel('Salary in $')
plt.show()
```

```python
salary_usa_by_devtype = salary_in_usa.dropna(subset=['DevType'])
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.set_theme(style='white', context='talk')
sns.barplot(data=salary_usa_pro_dev[:20], y='ConvertedCompYearly', x='LanguageHaveWorkedWith', hue='YearsCodePro', palette='pastel', orient='v')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(custom_format))
plt.xticks(rotation=90, size=8)
ax.legend(title='YearsCodePro', loc='upper left', bbox_to_anchor=(1, 1))
plt.xlabel('LanguageHaveWorkedWith')
plt.ylabel('Salary in $')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.set_theme(style='darkgrid', context='talk')
sns.boxplot(data=salary_usa_by_devtype[:40], x='DevType',
            y='ConvertedCompYearly',
            hue='DevType',)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(custom_format))
plt.xticks(rotation=90, size=8)
plt.xlabel('Developer Type')
plt.ylabel('Salary in $')
plt.show()
```

```python
emp_cat = list(df_employment_counts[df_employment_counts.values > 300].index)
emp_cat
```
